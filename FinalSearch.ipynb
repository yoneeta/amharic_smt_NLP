```python
import json
from difflib import SequenceMatcher
```

```python
FILE = "AbebeBesoBela statistics overall.json"#'BBC 2 statistics overall.json''ar_tr_pair_small.json' #data file with the translations
```

```python
def fileOpener():
    with open(FILE, 'r', encoding = 'utf-8') as f:
        gcorpus = json.load(f)
    return gcorpus

def sortmeout(e):
    
   # print(e.items())
    myshortarray=[]
    higher=""
    minHit=0.1
    for x,v in e.items():
        #print(minHit," ---- ",v)
        if  v > minHit:
            #print(v ," ----- ", minHit)
            minHit=v
            higher=x
        #else:
        #    print("")#minHit," some value ",v)
    #print("uuuu",higher,minHit)
    return higher

def charSequencSimilarity(str1,str2):
    
    s = SequenceMatcher(None, str1,str2)
    s.ratio()
    #print(jellyfish.damerau_levenshtein_distance(str1,str2))
    #print(jellyfish.jaro_distance(str1,str2))
    #print(jellyfish.levenshtein_distance(str1,str2))
    #print( s.ratio())

    return s.ratio()

corpTable=fileOpener()
def searchMe(sw):


    for x,u in corpTable.items():
        #print(x)
        #print(u)
        #if sw in x:
        #    print(sw+" i got "+x+" high value")
        #    #print(u)
        #    #print(sortmeout(u))
        #    return sortmeout(u)#x
        if len(sw) == len(x) :
            intersectoral=set(sw)&set(x)
            if sw == x:
                return sortmeout(u)
            else:
                if  len(intersectoral) >3 and len(intersectoral) >= round(len(sw)/2) and charSequencSimilarity(sw,x) > 0.3:

                        #print(sw,x)
                        return sortmeout(u)
        elif len(sw) > len(x) and x in sw:
            #print("o > i",sw,x)
            return sortmeout(u)
        elif len(sw) < len(x) and sw in x:
            #print("o < i",sw, x)
            return sortmeout(u)
```

```python
def breakDown(zi):
    y=[]
    #print(z)
    for searchW in zi:
        #print(searchW)
        y.append( searchMe(searchW))
    #print("---",y)
    return y

def alighneLanguageModel(en,am):
    #now lets work on structure
    #ens=en.split()
    #ams=am.split()
    #print(len(en))
    #now we assume the user input
    #actually it requires a POS tagging tool
    
    if len(am) == 3 and len(en) == 3:
        print("S O V",en)        
        fullSent=en[0]+" "+en[2]+" "+en[1]
        print(fullSent)
    elif len(am) == 2 and len(en) == 2:
        print("S V","O V")
        fullsent=en[0]+" "+en[1]
        print(fullsent)
    else:
        print(en[0])
        

z="ኳስ ተጫወተ"
zi=z.split()

yi=breakDown(zi)
#print(yi)
alighneLanguageModel(yi,zi)
#print(z)
```
